```python
import os
from PIL import Image
```

```python
input_folder = r"THYZ_2025_oturum2"
output_folder = r"Frames_jpg"
```

```python
for dosya_adi  in os.listdir(input_folder):
    if dosya_adi.lower().endswith(".webp"):
        dosya_yolu = os.path.join(input_folder, dosya_adi)

        jpg_adi = os.path.splitext(dosya_adi)[0] + ".jpg"
        jpg_yolu = os.path.join(output_folder, jpg_adi)

        try:
            with Image.open(dosya_yolu) as img:
                rgb_img = img.convert("RGB")
                rgb_img.save(jpg_yolu, "JPEG", quality=95)
            
            print(f"Çevrildi: {dosya_adi}")
        
        except Exception as e:
            print(f"Hata oluştu: {e}")

print("Dönüştürme tamamlandi.")
```
